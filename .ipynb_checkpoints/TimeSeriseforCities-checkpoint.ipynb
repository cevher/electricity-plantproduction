```python
from statsmodels.tsa.stattools import adfuller
from sklearn.preprocessing import MinMaxScaler
df = pd.DataFrame()

def check_stationarity(series, signif=0.05, name='', verbose=False):
    

    result = adfuller(series.values)
    print(f'    Augmented Dickey-Fuller Test on "{name}"', "\n   ", '-'*47)

    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

    if (result[1] <= 0.05) & (result[4]['5%'] > result[0]):
        print("\u001b[32mStationary\u001b[0m")
    else:
        print("\x1b[31mNon-stationary\x1b[0m")
        
def normalize(column):
    scaler = MinMaxScaler()
    data = column.to_numpy()
    data = data.reshape(-1, 1)
    scaler.fit(data)
    return scaler.transform(data)
```

```python
train1 = data1[:22]
test1 = data1[22:]
train2 = data2[:22]
test2 = data2[22:]
```

```python
predictions = []
def auto_arima(train1, train2):
    for column_tr1, column_tr2 in zip(train1.columns, train2.columns):
        x = train1[column_tr1].to_numpy().reshape(-1,1)
        y = train2[column_tr2].to_numpy().reshape(-1,1)
        arima = pm.auto_arima(x,y,error_action='ignore', trace=True,
                      suppress_warnings=True, maxiter=500,m=1,stationary=False,test='adf')

        # Print out summary information on the fit
        # print(arima.summary())
        # print(arima.params)
        X_test = test2[column_tr2].to_numpy().reshape(-1,1)
        pred=arima.predict(n_periods=2, X=X_test, alpha=0.05)
        predictions.append(pred)
        print(f"Prediction for {column_tr1}", pred)
                                      
    
```
